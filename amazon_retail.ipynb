```python
# Calculate total revenue
amazon['revenue'] = amazon['discounted_price']

# Calculate average order value
average_order_value = amazon['revenue'].mean()

# Calculate total number of sales
total_sales = amazon.shape[0]

print("Total Revenue: ₹", amazon['revenue'].sum())
print("Average Order Value: ₹", average_order_value)
print("Total Number of Sales:", total_sales)
```

```python
import pandas as pd

# Calculate total revenue
total_revenue = amazon['revenue'].sum()

# Calculate average order value
average_order_value = amazon['revenue'].mean()

# Calculate total number of sales
total_sales = amazon.shape[0]

# Create a DataFrame for the basic sales metrics
sales_metrics = pd.DataFrame({
    'Metric': ['Total Revenue', 'Average Order Value', 'Total Number of Sales'],
    'Value': [total_revenue, average_order_value, total_sales]
})

# Display the table
print(sales_metrics)
```

```python
# Analyze Product Performance based on Revenue
product_revenue = amazon.groupby('product_name')['revenue'].sum().sort_values(ascending=False)

# Identify Top-Selling Products based on Revenue
top_selling_products = product_revenue.head(5)

# Identify Top-Selling Categories based on Revenue (if available)
top_selling_categories = amazon.groupby('category')['revenue'].sum().sort_values(ascending=False).head(5)

# Identify Slow-Moving Items based on Revenue
slow_moving_items = product_revenue.tail(5)

# Display the results
print("Top-Selling Products based on Revenue:")
print(top_selling_products)
print("\nTop-Selling Categories based on Revenue:")
print(top_selling_categories)
print("\nProduct Performance based on Revenue:")
print(product_revenue.head(5))
print("\nSlow-Moving Items based on Revenue:")
print(slow_moving_items)
```

```python
# Analyze Product Performance based on Revenue
product_revenue = amazon.groupby('product_name')['revenue'].sum().sort_values(ascending=False)

# Identify Top-Selling Products based on Revenue
top_selling_products = product_revenue.head(5)

# Identify Top-Selling Categories based on Revenue
top_selling_categories = amazon.groupby('category')['revenue'].sum().sort_values(ascending=False).head(5)

# Create a bar chart for Top Products generating revenues
plt.figure(figsize=(8, 6))
plt.barh(top_selling_products.index, top_selling_products.values, color='skyblue')
plt.xlabel('Total Revenue')
plt.ylabel('Product Name')
plt.title('Top Products Generating Revenues')
plt.show()

# Create a bar chart for Top Categories and their revenues
plt.figure(figsize=(8, 6))
plt.barh(top_selling_categories.index, top_selling_categories.values, color='lightcoral')
plt.xlabel('Total Revenue')
plt.ylabel('Category')
plt.title('Top Categories and their Revenues')
plt.show()

# Insights
print("Insights:")
print("Top Products Generating Revenues:")
print(top_selling_products)
print("\nTop Categories and Their Revenues:")
print(top_selling_categories)
```

```python
# Convert price-related columns to numeric
amazon['discounted_price'] = pd.to_numeric(amazon['discounted_price'].astype(str).str.replace('$', '').str.replace(',', ''), errors='coerce')
amazon['actual_price'] = pd.to_numeric(amazon['actual_price'].astype(str).str.replace('$', '').str.replace(',', ''), errors='coerce')
amazon['discount_percentage'] = pd.to_numeric(amazon['discount_percentage'].astype(str).str.replace('%', ''), errors='coerce')

# Drop rows with missing values in any of the required columns
amazon = amazon.dropna(subset=['discounted_price', 'actual_price', 'discount_percentage'])

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(amazon['actual_price'], amazon['discounted_price'], c=amazon['discount_percentage'], cmap='coolwarm', alpha=0.7)
plt.colorbar(label='Discount Percentage')
plt.xlabel('Actual Price ($)')
plt.ylabel('Discounted Price ($)')
plt.title('Relationship between Actual Price and Discounted Price')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Calculate Pearson correlation coefficient
pearson_corr_actual_discounted = np.corrcoef(amazon['actual_price'], amazon['discounted_price'])[0, 1]
pearson_corr_discount_percentage_actual = np.corrcoef(amazon['discount_percentage'], amazon['actual_price'])[0, 1]
pearson_corr_discount_percentage_discounted = np.corrcoef(amazon['discount_percentage'], amazon['discounted_price'])[0, 1]

print("Pearson Correlation Coefficients:")
print("Actual Price vs. Discounted Price:", pearson_corr_actual_discounted)
print("Discount Percentage vs. Actual Price:", pearson_corr_discount_percentage_actual)
print("Discount Percentage vs. Discounted Price:", pearson_corr_discount_percentage_discounted)
```

<span style="color:blue">**Actual Price vs. Discounted Price (0.9619):**</span>  
There's a strong positive correlation (0.9619) between the listed actual price and discounted price on Amazon, reflecting consistent and transparent pricing practices.

<span style="color:green">**Discount Percentage vs. Actual Price (-0.1175):**</span>  
While higher-priced items tend to receive lower discount percentages, the effect is marginal, indicating minimal influence on pricing strategies.

<span style="color:red">**Discount Percentage vs. Discounted Price (-0.2420):**</span>  
There's a modest negative correlation (-0.2420) between discount percentage and discounted price, suggesting a balanced approach to pricing adjustments to maintain competitiveness and profitability on Amazon.

```python
# Count the occurrences of each category
category_counts = amazon['category'].value_counts()

# Select top 10 categories
top_10_categories = category_counts.head(10)
other_categories_count = category_counts[10:].sum()

# Combine remaining categories into 'Others'
top_10_categories['Others'] = other_categories_count

# Plot the pie chart
plt.figure(figsize=(10, 6))
plt.pie(top_10_categories, labels=top_10_categories.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Category Types Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

**<font color='blue'>Insight</font>**

The dominance of 'Electronics' as the top category on Amazon highlights robust consumer demand, substantial revenue contribution, and competitive prospects within this segment.

```python
# Count the occurrences of each category
category_counts = amazon['category'].value_counts()

# Select top 5 categories
top_5_categories = category_counts.head(5)
other_categories_count = category_counts[5:].sum()

# Combine remaining categories into 'Others'
top_5_categories['Others'] = other_categories_count

# Plot the pie chart and legend box side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot the pie chart
pie, _ = ax1.pie(top_5_categories, labels=[''] * len(top_5_categories), startangle=140)
ax1.set_title('Top 5 Category Types Distribution')
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Create a legend box
legend_handles = [(plt.Rectangle((0,0),1,1, color=pie[i].get_facecolor()), category) for i, category in enumerate(top_5_categories.index)]
ax2.legend([handle[0] for handle in legend_handles], top_5_categories.index, loc='center', title='Category Legend', title_fontsize='x-large', fontsize='large', fancybox=True, facecolor='lightgray')
ax2.axis('off')  # Hide axis

plt.tight_layout()  # Adjust layout to prevent overlap

plt.show()
```

```python
# Define function to aggregate categories
def aggregate_categories(category):
    if 'Kitchen' in category or 'Home' in category:
        return 'Home & Kitchen'
    elif 'USB' in category or 'Computers' in category or 'Accessories' in category:
        return 'Electronics'
    elif 'Office Products' in category:
        return 'Office Products'
    else:
        return category

# Apply function to aggregate categories
amazon['category'] = amazon['category'].apply(aggregate_categories)

# Count the occurrences of each category
category_counts = amazon['category'].value_counts()

# Select top 5 categories
top_5_categories = category_counts.head(5)
other_categories_count = category_counts[5:].sum()

# Combine remaining categories into 'Others'
top_5_categories['Others'] = other_categories_count

# Plot the pie chart and legend box side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot the pie chart
pie, _ = ax1.pie(top_5_categories, labels=[''] * len(top_5_categories), startangle=140)
ax1.set_title('Top 5 Category Types Distribution')
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Create a legend box
legend_handles = [(plt.Rectangle((0,0),1,1, color=pie[i].get_facecolor()), category) for i, category in enumerate(top_5_categories.index)]
ax2.legend([handle[0] for handle in legend_handles], top_5_categories.index, loc='center', title='Category Legend', title_fontsize='x-large', fontsize='large', fancybox=True, facecolor='lightgray')
ax2.axis('off')  # Hide axis

plt.tight_layout()  # Adjust layout to prevent overlap

plt.show()
```

# [Reviews & Ratings](#4)


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot of Ratings
plt.figure(figsize=(8, 6))
sns.countplot(data=amazon, x='rating', palette='viridis')
plt.title('Count of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Convert 'rating' column to numeric type
amazon['rating'] = pd.to_numeric(amazon['rating'], errors='coerce')

# Round the ratings to the nearest integer
amazon['rounded_rating'] = np.ceil(amazon['rating'])

# Countplot of rounded ratings
plt.figure(figsize=(8, 6))
sns.countplot(data=amazon, x='rounded_rating', palette='viridis')
plt.title('Count of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Countplot of reviews for each rating score
plt.figure(figsize=(10, 6))
sns.countplot(data=amazon, x='rating', palette='viridis')
plt.title('Count of Reviews for Each Rating Score')
plt.xlabel('Rating Score')
plt.ylabel('Count of Reviews')
plt.xticks(rotation=45)
plt.show()
```
